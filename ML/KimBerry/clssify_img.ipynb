```python
#From d2l book
from d2l import tensorflow as d2l
import matplotlib.pyplot as plt

d2l.use_svg_display()
mnist_train, mnist_test = tf.keras.datasets.fashion_mnist.load_data()

#questtin: how can we change the normal images into these type of data (with vector)?

def get_fashion_mnist_labels(labels):  #@save
    """Return text labels for the Fashion-MNIST dataset."""
    text_labels = [
        't-shirt', 'trouser', 'pullover', 'dress', 'coat', 'sandal', 'shirt',
        'sneaker', 'bag', 'ankle boot']
    return [text_labels[int(i)] for i in labels]

def show_images(imgs, num_rows, num_cols, titles=None, scale=1.5):  #@save
    """Plot a list of images."""
    figsize = (num_cols * scale, num_rows * scale)
    _, axes = d2l.plt.subplots(num_rows, num_cols, figsize=figsize) # _ means that we wont use it?
    print(axes)
    axes = axes.flatten() #make any matrices into a column vector
    
    for i, (ax, img) in enumerate(zip(axes, imgs)):
        
        ax.imshow(img.numpy())
        ax.axes.get_xaxis().set_visible(False)
        ax.axes.get_yaxis().set_visible(False)
        if titles:
            ax.set_title(titles[i])
    return axes
#How can we change the data X into the img?
X = tf.constant(mnist_train[0][18:19])
y = tf.constant(mnist_train[1][18:19])
show_images(X, 2, 1, titles=get_fashion_mnist_labels(y)); 
#print(X)
print(X.shape)
print(X)
#กลับไปดูไฟล์ Detext_Malaria_secondAttempt ลองรัน แล้วจะเข้าใจเอง
# mnist_train's type is tuple
```
